# Exploratory Data Analysis ABR Salmonella enteriditis

> The analysis of antibiotic resistance in Salmonella isolates provides critical insights into the effectiveness of treatment options available for bacterial infections. By examining the data collected over several years and across different regions, we aim to understand patterns of resistance that can inform healthcare practices and policy decisions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the uploaded CSV files
resistance_percentage_df = pd.read_csv('salmonella_antibiotic_resistance_percentage.csv')
isolate_df = pd.read_csv('salmonella_isolate_data_cleaned.csv')
```

```python
# Display the first few rows of each DataFrame to understand their structure
resistance_percentage_df.head()
```

```python
isolate_df.head(10)
```

```python
# 1. Countplot for Data Year
plt.figure(figsize=(10, 6))
sns.countplot(data=isolate_df, x='Data Year', color=sns.color_palette("viridis")[0])
plt.title('Counts for Data Year', fontsize=16)
plt.xlabel('Data Year', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Region Summary
region_summary = isolate_df['Region Name'].value_counts().reset_index()
region_summary.columns = ['Region Name', 'Count']

# Define the custom order for regions
region_order = [f'Region {i}' for i in range(1, 11)]  # Adjust based on the actual number of regions

# Convert Region Name to a categorical type with the defined order
region_summary['Region Name'] = pd.Categorical(region_summary['Region Name'], categories=region_order, ordered=True)
region_summary = region_summary.sort_values('Region Name')  # Sort by custom order

# Plot Region Summary
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Region Name', data=region_summary, color=sns.color_palette("viridis")[1])
plt.title('Region Counts', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Region Name', fontsize=14)
plt.show()
```

```python
# Age Group Summary
age_group_summary = isolate_df['Age Group'].value_counts().reset_index()
age_group_summary.columns = ['Age Group', 'Count']

# Define the correct custom order for age groups
age_group_order = ['0-4', '5-9', '10-19', '20-29', '30-39', 
                   '40-49', '50-59', '60-69', '70-79', '80+']

# Convert Age Group to a categorical type with the defined order
age_group_summary['Age Group'] = pd.Categorical(age_group_summary['Age Group'], categories=age_group_order, ordered=True)
age_group_summary = age_group_summary.sort_values('Age Group')  # Sort by custom order

# Plot Age Group Summary
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Age Group', data=age_group_summary, color=sns.color_palette("viridis")[2])
plt.title('Age Group Counts', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Age Group', fontsize=14)
plt.show()
```

```python
# 3. Specimen Source Summary (optional)
specimen_source_summary = isolate_df['Specimen Source'].value_counts().reset_index()
specimen_source_summary.columns = ['Specimen Source', 'Count']
specimen_source_summary = specimen_source_summary.sort_values(by='Count', ascending=False)  # Sort by Count

# Plot Specimen Source Summary
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Specimen Source', data=specimen_source_summary, color=sns.color_palette("viridis")[3])
plt.title('Specimen Source Counts', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Specimen Source', fontsize=14)
plt.show()
```

```python
# Define the columns that end with 'Concl'
concl_columns = [col for col in isolate_df.columns if col.endswith('Concl')]

# Create a figure for the counts of each resistance level (S, R, I)
resistance_levels = {
    'S': 'Susceptible',
    'R': 'Resistant',
    'I': 'Intermediate'
}

# Initialize a dictionary to store counts for each resistance level across all 'Concl' columns
combined_counts = {level: [] for level in resistance_levels.keys()}

# Loop through each 'Concl' column to accumulate counts for S, R, I
for column in concl_columns:
    for level in resistance_levels.keys():
        # Exclude 'X' values
        count = isolate_df[column].value_counts().get(level, 0)  # Get count or 0 if not present
        combined_counts[level].append(count)

# Create a DataFrame for plotting
combined_counts_df = pd.DataFrame(combined_counts, index=concl_columns)

# Plot for Susceptible (S)
plt.figure(figsize=(8, 5))
sns.barplot(x=combined_counts_df.index, y=combined_counts_df['S'], color=sns.color_palette("viridis")[0])
plt.title('Counts for Resistance Level: Susceptible', fontsize=16)
plt.xlabel('Antibiotic', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout
plt.show()

# Plot for Resistant (R)
plt.figure(figsize=(8, 5))
sns.barplot(x=combined_counts_df.index, y=combined_counts_df['R'], color=sns.color_palette("viridis")[1])
plt.title('Counts for Resistance Level: Resistant', fontsize=16)
plt.xlabel('Antibiotic', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout
plt.show()

# Plot for Intermediate (I)
plt.figure(figsize=(8, 5))
sns.barplot(x=combined_counts_df.index, y=combined_counts_df['I'], color=sns.color_palette("viridis")[2])
plt.title('Counts for Resistance Level: Intermediate', fontsize=16)
plt.xlabel('Antibiotic', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# Create a count plot for the MAR_Level column
plt.figure(figsize=(8, 5))
sns.countplot(data=isolate_df, x='AR Level', color=sns.color_palette("viridis")[0])
plt.title('Counts of Antibiotic Resistance Level', fontsize=16)
plt.xlabel('AR Level', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Group by 'Data Year' and 'AR Level' to get the counts, using observed=True
ar_year_counts = isolate_df.groupby(['Data Year', 'AR Level'], observed=True).size().reset_index(name='Count')

# Create a line plot for AR Level counts over Data Year
plt.figure(figsize=(9, 6))

# Generate a viridis color palette based on the number of unique AR Levels
viridis_colors = sns.color_palette("viridis", n_colors=len(ar_year_counts['AR Level'].unique()))

# Plotting each AR Level with the generated colors
for i, ar_level in enumerate(sorted(ar_year_counts['AR Level'].unique())):
    level_data = ar_year_counts[ar_year_counts['AR Level'] == ar_level]
    
    # Check if level_data has data to plot
    if not level_data.empty:
        plt.plot(level_data['Data Year'], level_data['Count'], marker='o', label=f'AR Level {ar_level}', color=viridis_colors[i])

# Adding titles and labels
plt.title('AR Level Counts (0-8) Over Years', fontsize=16)
plt.xlabel('Data Year', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title='AR Level')
plt.show()
```

```python
# Group by 'Data Year' and 'AR Level' to get the counts
ar_year_counts = isolate_df.groupby(['Data Year', 'AR Level'], observed=True).size().reset_index(name='Count')

# Create a new DataFrame to hold the summed counts
summary_counts = ar_year_counts.groupby('Data Year').agg(
    Susceptible=('Count', lambda x: x[ar_year_counts['AR Level'] == 0].sum()),  # Sum counts where AR Level is 0
    Resistant=('Count', lambda x: x[ar_year_counts['AR Level'] > 0].sum())   # Sum counts where AR Level is greater than 0
).reset_index()

# Create a line plot for Susceptible vs Resistant counts over Data Year
plt.figure(figsize=(12, 6))

# Get the Viridis color palette
viridis_colors = sns.color_palette("viridis", n_colors=2)

# Plotting the susceptible counts
sns.lineplot(data=summary_counts, x='Data Year', y='Susceptible', marker='o', label='Susceptible (AR Level 0)', color=viridis_colors[0])
# Plotting the resistant counts
sns.lineplot(data=summary_counts, x='Data Year', y='Resistant', marker='o', label='Resistant (AR Level 1+)', color=viridis_colors[1])

# Adding titles and labels
plt.title('Susceptible vs Resistant Counts Over Years', fontsize=16)
plt.xlabel('Data Year', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.legend(title='AR Level')
plt.tight_layout()
plt.show()
```

```python
# Step 1: Select relevant columns
concl_columns = [col for col in isolate_df.columns if col.endswith('Concl')]
selected_columns = ['Data Year', 'Region Name', 'Age Group', 'Specimen Source'] + concl_columns

# Create a copy of the filtered DataFrame to avoid SettingWithCopyWarning
filtered_df = isolate_df[selected_columns].copy()

# Step 2: Encode categorical variables using Label Encoding
label_encoder = LabelEncoder()

# Encode categorical columns using .loc to avoid the warning
filtered_df.loc[:, 'Region Name'] = label_encoder.fit_transform(filtered_df['Region Name'])
filtered_df.loc[:, 'Age Group'] = label_encoder.fit_transform(filtered_df['Age Group'])
filtered_df.loc[:, 'Specimen Source'] = label_encoder.fit_transform(filtered_df['Specimen Source'])

# For 'Concl' columns, use label encoding
for col in concl_columns:
    filtered_df.loc[:, col] = label_encoder.fit_transform(filtered_df[col])

# Step 3: Calculate the correlation matrix
correlation_matrix = filtered_df.corr()

# Step 4: Plot the correlation map
plt.figure(figsize=(16, 11))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='viridis', square=True, linewidths=.5)
plt.title('Correlation Map', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Select the antibiotic resistance percentage columns
resistance_percentage_columns = [col for col in resistance_percentage_df.columns if 'Resistance Percentage' in col]

# Loop through each antibiotic resistance column and create separate plots
for antibiotic in resistance_percentage_columns:
    plt.figure(figsize=(8, 5))
    
    # Use the entire range of years for the x-axis
    sns.lineplot(data=resistance_percentage_df, x='Data Year', y=antibiotic, marker='o', color=sns.color_palette("viridis")[0])
    
    # Adding titles and labels
    plt.title(f'Antibiotic Resistance Percentage Over Time: {antibiotic}', fontsize=16)
    plt.xlabel('Data Year', fontsize=14)
    plt.ylabel('Resistance Percentage', fontsize=14)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
    plt.xlim(resistance_percentage_df['Data Year'].min(), resistance_percentage_df['Data Year'].max())  # Ensure all years are included
    plt.tight_layout()
    plt.show()
```

```python
# You can directly use the antibiotic resistance columns
heatmap_data = resistance_percentage_df.groupby('Region Name').mean()[[
    'AMP Resistance Percentage',
    'AUG Resistance Percentage',
    'AXO Resistance Percentage',
    'CHL Resistance Percentage',
    'CIP Resistance Percentage',
    'COT Resistance Percentage',
    'FIS Resistance Percentage',
    'FOX Resistance Percentage',
    'GEN Resistance Percentage',
    'NAL Resistance Percentage',
    'STR Resistance Percentage',
    'TET Resistance Percentage',
    'Total Resistance Percentage'
]].reset_index()

# Set the index for the heatmap
heatmap_data.set_index('Region Name', inplace=True)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='viridis', linewidths=.5)
plt.title('Antibiotic Resistance Levels by Region', fontsize=16)
plt.xlabel('Antibiotic', fontsize=14)
plt.ylabel('Region', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Melt the DataFrame to long format for easier plotting
melted_df = resistance_percentage_df.melt(
    id_vars=['Data Year', 'Region Name'],
    value_vars=[
        'AMP Resistance Percentage',
        'AUG Resistance Percentage',
        'AXO Resistance Percentage',
        'CHL Resistance Percentage',
        'CIP Resistance Percentage',
        'COT Resistance Percentage',
        'FIS Resistance Percentage',
        'FOX Resistance Percentage',
        'GEN Resistance Percentage',
        'NAL Resistance Percentage',
        'STR Resistance Percentage',
        'TET Resistance Percentage'
    ],
    var_name='Antibiotic',
    value_name='Resistance Percentage'
)

# Create a grid of subplots
regions = melted_df['Region Name'].unique()
num_regions = len(regions)
num_antibiotics = len(melted_df['Antibiotic'].unique())
fig, axes = plt.subplots(num_regions, 1, figsize=(12, 5 * num_regions), sharex=True)

# Plot each region
for ax, region in zip(axes, regions):
    region_data = melted_df[melted_df['Region Name'] == region]
    sns.lineplot(data=region_data, x='Data Year', y='Resistance Percentage', hue='Antibiotic', ax=ax, marker='o')
    ax.set_title(f'Antibiotic Resistance Over Time in {region}', fontsize=16)
    ax.set_ylabel('Resistance Percentage', fontsize=14)

plt.xlabel('Year', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Antibiotic', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Summary of Visualizations

> - **Count Plot for Data Year**: This plot shows the distribution of samples across different years in the dataset. It helps identify trends in data collection over time.
>
> - **Resistance Percentages**: A series of plots displaying the percentage of resistance for various antibiotics. This helps visualize trends in antibiotic resistance over the years and across regions.
>
> - **Heatmaps**: Heatmaps were generated to show correlations between different antibiotic resistance percentages, allowing for an understanding of how resistance to one antibiotic may relate to another.
>
> - **Resistance Levels**: Analysis of resistance levels, helping to categorize and visualize how many samples fall into different levels of antibiotic resistance.


## Conclusions from EDA

> - **Temporal Trends**: The data collection spans multiple years, revealing trends in antibiotic resistance. Notably, certain antibiotics show increasing resistance percentages over time, suggesting a potential public health concern.
>
> - **Regional Variability**: Different regions exhibit variability in resistance levels, indicating that geographical factors may influence antibiotic effectiveness. This information can guide targeted interventions in regions with higher resistance rates.
>
> - **Correlation Between Antibiotics**: The heatmaps indicate varying degrees of correlation between resistance to different antibiotics. Strong correlations could suggest shared mechanisms of resistance or cross-resistance, which should be further investigated.
>
> - **Resistance Levels**: The analysis of resistance levels highlights the need for continuous monitoring of antibiotic susceptibility patterns. The categorization into different resistance levels allows for better clinical decision-making regarding treatment options.
